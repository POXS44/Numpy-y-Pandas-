##### b. Estructura de un cuaderno de jupyter notebook
##### Como esta dividido un cuaderno de jupyter notebook ?
Los mens se utilizan para interactuar con el cuaderno Jupyter. La barra de men est presente en la parte superior del  cuaderno con la cual podemos realizar diferentes acciones como crear cuaderno insertar celda vista interrumpir nucleo un sin numero de acciones .
 
#####  Cmo ejecutar cdigo?
 Ejecutar una celda significa ejecutar el contenido de la celda. Ver una lista de opciones debajo de la barra de men. El botn Ejecutar est presente entre las opciones. La forma alternativa de ejecutar una celda es, en su teclado, simplemente presione shift+enter
 
 
##### Archivo
   
##### De las opciones enumeradas en el men de archivo, podemos entender que podemos hacer las siguientes cosas con el men de archivo.

- Crea un nuevo cuaderno.
- Abra uno existente.
- Cambiar el nombre de un cuaderno.
- Guarde un cuaderno.
- Con la ayuda de la opcin de punto de control, puede crear un punto de control al que puede retroceder cuando sea necesario.
##### Editar
El men de edicin sirve para cortar, copiar, pegar y reordenar las celdas. Tambin proporciona la funcionalidad para eliminar, fusionar o dividir celdas. Puede notar que algunas de las opciones en el men de edicin estn atenuadas. Significa que las opciones no son aplicables a la celda actual. Por ejemplo, podemos insertar una imagen en una celda de Markdown, pero en una celda de cdigo, no podemos. Si ve una opcin atenuada, cambie el tipo de celda y vea si est disponible para usar ahora. El men Celda explica cmo cambiar el tipo de celda.
##### Ver
El men de vista tiene que ver con cambiar la visibilidad de la celda. Proporciona la funcionalidad para alternar el encabezado y la barra de herramientas. Tambin puede alternar los nmeros de lnea dentro de la celda y jugar con la barra de herramientas de la celda.
##### Formato en celda
En esta parte podemos ver en que formato queremos ir creando nuestra celda.

##### recuadro
Si desea agregar una nueva celda arriba o la celda seleccionada, la opcin de insercin es la indicada.   
##### Telfono mvil
Como se ve en el men Editar, si desea cambiar el tipo de celda, debe ir al men de la celda. Adems, el men de celdas permite al usuario ejecutar una celda seleccionada, un grupo de celdas arriba o abajo de las celdas seleccionadas e incluso todas las celdas. Suponga que desea compartir su celda con otras personas, es posible que deba borrar la salida de la celda para que otras personas puedan ejecutarse. El men de la celda ofrece la opcin de borrar la salida actual as como las salidas.
##### Ayuda
El men de ayuda proporciona al usuario una gran cantidad de materiales de referencia, un recorrido por la interfaz de usuario y atajos de teclado.
##### Widgets
Los widgets no son ms que widgets de Javascript que los usuarios pueden agregar a sus celdas para hacer que la celda sea ms dinmica. Tambin puede guardar y borrar el widget con el men de widgets.
